```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list=[10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list=[20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list, weight_list,"y") #y-> yellow, g, b, r
plt.xlabel("Age")
plt.ylabel("weight")
```

```python
#numpy
```

```python
np_age_list=np.array(age_list)
```

```python
np_weight_list=np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"b")
plt.xlabel("AGE")
plt.ylabel("WEIGHT")
```

```python
numpy_arr1 = np.linspace(0,10,20) 
#sfrdan ona kadar 20 tane eit aralklarla say retir
```

```python
numpy_arr2 = numpy_arr1**3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2,"pink")
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2,numpy_arr1,"b--")
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
```
